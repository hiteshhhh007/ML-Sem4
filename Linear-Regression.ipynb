# Import Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

# Create Dataset

```python
np.random.seed(42)  # For reproducibility
X = np.random.rand(100, 1) * 10  # 100 random values from 0 to 10
y = 2.5 * X + np.random.randn(100, 1) * 2  # y = 2.5 * X + some noise
```

```python
df = pd.DataFrame({"X": X.flatten(), "y": y.flatten()})
```

# Scatter Plot

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="X", y="y", data=df)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter Plot of X vs. y")
plt.show()
```

# Regressor Line

```python
# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Get predictions
y_pred = model.predict(X)

# Scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.scatterplot(x="X", y="y", data=df, label="Data Points")
plt.plot(X, y_pred, color="red", label="Regressor Line")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Regression Line on Scatter Plot")
plt.legend()
plt.show()
```

# Predicted ð‘¦


```python
print("Predicted y values:")
print(y_pred.flatten())
```

# Prediction for New Values

```python
# New values for prediction
X_new = np.array([[11], [12], [13]])  # New input values

# Predict outcomes
y_new_pred = model.predict(X_new)

print("\nPredictions for new values:")
for i, val in enumerate(X_new):
    print(f"X = {val[0]} -> Predicted y = {y_new_pred[i][0]}")
```

# Squared Error and Mean Squared Error

```python
# Squared error
squared_errors = (y - y_pred) ** 2

# Mean squared error
mse = mean_squared_error(y, y_pred)

print("\nSquared Errors:")
print(squared_errors.flatten())

print("\nMean Squared Error:")
print(mse)
```

# Absolute Error and Mean Absolute Error

```python
# Absolute error
absolute_errors = np.abs(y - y_pred)

# Mean absolute error
mae = mean_absolute_error(y, y_pred)

print("\nAbsolute Errors:")
print(absolute_errors.flatten())

print("\nMean Absolute Error:")
print(mae)
```

# R-squared and Adjusted R-squared

```python
# R-squared
r_squared = r2_score(y, y_pred)

# Adjusted R-squared
n = len(y)  # number of observations
p = 1  # number of predictors
adjusted_r_squared = 1 - (1 - r_squared) * (n - 1) / (n - p - 1)

print("\nR-squared:")
print(r_squared)

print("\nAdjusted R-squared:")
print(adjusted_r_squared)
```
